#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('sinasc_ro.csv')
df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
grafico_df = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
for coluna in grafico_df:
    plt.figure(figsize=(4, 2))
    df[coluna].value_counts().sort_index().plot(kind='bar')
    plt.title(coluna)
    plt.xlabel('Frequência')
    plt.ylabel('Quantidade')
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
avg_apgar5_by_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()
avg_apgar5_by_consultas
plt.figure(figsize=(5, 3))
avg_apgar5_by_consultas.plot(kind='bar', color='blue')
plt.title('APGAR5 Médio por Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.xticks(rotation=45)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
#SEXO
plt.subplot(2,2,1)
sns.countplot(data=df,x='SEXO')
plt.title('Frequência de Sexo')
plt.xlabel('Sexo')
plt.ylabel('Frequência')
#GESTACAO
plt.subplot(2,2,2)
sns.countplot(data=df,x='GESTACAO')
plt.title('Frequência de Gestação')
plt.xlabel('Duração da Gestação')
plt.xticks(rotation=45)
plt.ylabel('Frequência')
#ESTCIVMAE
plt.subplot(2,2,3)
sns.countplot(data=df,x='ESTCIVMAE')
plt.title('Frequência do Estado civil da Mãe')
plt.xlabel('Estado civil da Mãe')
plt.ylabel('Frequência')
#CONSULTAS
plt.subplot(2,2,4)
sns.countplot(data=df,x='CONSULTAS')
plt.title('Frequência de CONSULTAS')
plt.xlabel('Quantidade de CONSULTAS')
plt.ylabel('Frequência')
# Ajuste de espaçamento entre os gráficos
plt.tight_layout()
#painel de gráficos
plt.show()
```
